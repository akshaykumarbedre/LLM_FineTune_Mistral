```python
from datasets import load_dataset
import pandas as pd
import json
import numpy as np # Import numpy

# Define a custom JSONEncoder to handle NumPy types
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NumpyEncoder, self).default(obj)

# Load the OpenAssistant/oasst1 dataset
print("Loading OpenAssistant/oasst1 dataset...")
oasst_dataset = load_dataset("OpenAssistant/oasst1", split="train")
print("Dataset loaded.")

# Convert to a Pandas DataFrame for easier manipulation
df = df_eng

# Helper function to reconstruct conversation paths
def get_conversation_path(message_id, df_messages):
    path = []
    current_message = df_messages[df_messages['message_id'] == message_id]
    if current_message.empty: # Handle case where message_id might not be found (e.g., due to filtering)
        return []

    current_message = current_message.iloc[0]
    
    temp_path_reversed = []
    while pd.notna(current_message['parent_id']):
        temp_path_reversed.append(current_message)
        parent_id = current_message['parent_id']
        parent_msg = df_messages[df_messages['message_id'] == parent_id]
        if parent_msg.empty: # Parent not found in this subset, stop
            break
        current_message = parent_msg.iloc[0]
    temp_path_reversed.append(current_message) 

    path = temp_path_reversed[::-1]
    return path

# List to store our transformed data
openai_formatted_conversations = []

# Process each unique conversation tree
unique_tree_ids = df['message_tree_id'].unique()

print(f"Processing {len(unique_tree_ids)} conversation trees...")

for tree_id in unique_tree_ids:
    tree_messages_df = df[df['message_tree_id'] == tree_id].sort_values(by='created_date')
    
    # Get all parent_ids within this tree to identify non-leaf nodes
    parents_in_tree = set(tree_messages_df['parent_id'].dropna().unique())
    
    # Leaf messages are those whose message_id is not in the set of parents_in_tree
    leaf_messages = tree_messages_df[~tree_messages_df['message_id'].isin(parents_in_tree)]
    
    for _, leaf_message in leaf_messages.iterrows():
        conversation_path = get_conversation_path(leaf_message['message_id'], tree_messages_df)
        
        if not conversation_path: # Skip if path reconstruction failed (e.g. root not found)
            continue

        openai_messages = []
        
        # Collect relevant metadata for the entire conversation
        # Ensure that `labels` are converted to a standard Python dict/list if they are a Series or NumPy array
        labels_data = leaf_message['labels']
        if isinstance(labels_data, pd.Series):
            # Convert Pandas Series (which might contain numpy types) to a standard dict
            # This handles cases where labels might be dicts-of-lists-of-numpy-types
            labels_data = labels_data.apply(lambda x: x.tolist() if isinstance(x, np.ndarray) else x).to_dict()
        elif isinstance(labels_data, dict):
            # Recursively process dict to ensure no numpy types remain
            labels_data = json.loads(json.dumps(labels_data, cls=NumpyEncoder))
        
        detoxify_data = leaf_message['detoxify']
        if isinstance(detoxify_data, pd.Series):
             detoxify_data = detoxify_data.apply(lambda x: x.tolist() if isinstance(x, np.ndarray) else x).to_dict()
        elif isinstance(detoxify_data, dict):
            detoxify_data = json.loads(json.dumps(detoxify_data, cls=NumpyEncoder))

        conversation_metadata = {
            "message_tree_id": tree_id,
            "final_message_id": leaf_message['message_id'],
            "final_message_lang": leaf_message['lang'],
            "final_message_role": leaf_message['role'],
            "final_message_rank": int(leaf_message['rank']) if pd.notna(leaf_message['rank']) else None, # Convert to int
            "final_message_review_result": bool(leaf_message['review_result']) if pd.notna(leaf_message['review_result']) else None, # Convert to bool
            "final_message_synthetic": bool(leaf_message['synthetic']) if pd.notna(leaf_message['synthetic']) else None, # Convert to bool
            "final_message_detoxify": detoxify_data,
            "final_message_labels": labels_data,
            "final_message_emojis": leaf_message['emojis'], # This field might also need careful handling if it contains numpy types
            "final_message_tree_state": leaf_message['tree_state']
        }

        for msg in conversation_path:
            role = "user" if msg['role'] == "prompter" else "assistant"
            
            message_obj = {
                "role": role,
                "content": msg['text']
            }
            openai_messages.append(message_obj)
        
        openai_formatted_conversations.append({
            "messages": openai_messages,
        })

print(f"Generated {len(openai_formatted_conversations)} OpenAI-like conversations.")

# --- Example of how to use the generated data ---

# Print the first few entries
for i, conv_data in enumerate(openai_formatted_conversations[:5]):
    print(f"\n--- Conversation {i+1} ---")
    print("OpenAI Chat Format Messages:")
    # Use the custom encoder when printing for consistency
    print(json.dumps(conv_data['messages'], indent=2, ensure_ascii=False, cls=NumpyEncoder)) 
    print("\nOASST1 Metadata for this Conversation:")
    #print(json.dumps(conv_data['oasst_conversation_metadata'], indent=2, ensure_ascii=False, cls=NumpyEncoder))

# To save this data to a JSONL file (one JSON object per line)
# Note: Using .jsonl is generally better for large datasets as it's streamable.
```
